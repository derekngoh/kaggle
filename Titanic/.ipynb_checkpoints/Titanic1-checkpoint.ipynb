# Feature Engineering

```python
#Add features to better interpret data and see relationships better
###### Remove fare outliers and cluster fare to identify suitable fare groups,
###### which might indirectly tell / relate socio-economic class and potential cabin location / quality. (window / no window etc)

###### Group age groups into ranges

###### Identify potential family relation, family size, unaccompanied minor / teens
###### Split family name, group ticket / fares, embarkation

###### Impute missing age values and others, considering multiple features, e.g. embarkation, family relation, sibsp, parch etc

###### Check cabin is properly assigned; alpha only or alpha num
```

```python
import pandas as pd
import numpy as np
```

```python
titanic_raw_train = pd.read_csv("./titanic_data_raw/train.csv")
```

```python
print("Length of df = ", len(titanic_raw_train))
titanic_raw_train.head()
```

### Add family name column

```python
#create family name column
titanic_raw_train["LastName"] = titanic_raw_train["Name"].str.split(",", n=1, expand=True)[0]
titanic_raw_train.tail()
```

### Add age group column

```python
#create age range dataframe for mapping
age_map = {"lower_limit":[0.0,3.00001,7.00001,14.00001,18.00001,25.00001,35.00001,45.00001,55.00001,65.00001,75.00001], 
           "upper_limit":[3.0,7.0,14.0,18.0,25.0,35.0,45.0,55.0,65.0,75.0,150.0]}

age_map_df = pd.DataFrame(data=age_map)
age_map_df.reset_index(inplace=True)
age_map_df=age_map_df.rename(columns={'index':'age_grp'}) 
age_map_df['age_grp'] = age_map_df['age_grp'].astype('category')
age_map_df.head()
```

```python
#join the map to main df and retain only if age conditions match to get age_range group
titanic_raw_train = titanic_raw_train.merge(age_map_df, how="cross")
```

```python
titanic_raw_train = titanic_raw_train[(titanic_raw_train['Age']<=titanic_raw_train['upper_limit']) & (titanic_raw_train['Age']>=titanic_raw_train['lower_limit'])]
titanic_raw_train.drop(['lower_limit', 'upper_limit'], axis=1, inplace=True)
```

### Add cabin info columns

```python
# check cabin is properly assigned. Fill NaN cabin to 0 for proper processing.
titanic_raw_train["Cabin"] = titanic_raw_train["Cabin"].replace(np.nan,"not_assigned")
titanic_raw_train["Cabin"] = titanic_raw_train["Cabin"].astype(str)
titanic_raw_train["cabin_proper"] = list(map(lambda x: (not x.isalpha()) & (x.strip().isalnum()), titanic_raw_train['Cabin']))
```

```python
titanic_raw_train["multiple_cabin"] = list(map(lambda x: len(x)>1, titanic_raw_train["Cabin"].str.split(" "))) 
```

### Add potential family relationship

```python
#is potential family when they have the following in common
### same surname
### same embarkation
### same ticket
### same fare

titanic_raw_train['family_number'] = titanic_raw_train.groupby(['LastName', 'Ticket', 'Fare', 'Pclass', 'Embarked']).ngroup()
```

```python
#count family member. greater than 1 means a family, one means single.
titanic_raw_train['family_memb_count'] = titanic_raw_train.groupby(['family_number'])['family_number'].transform('count')
```

```python
titanic_raw_train["has_family"] = list(map(lambda x: (x>1), titanic_raw_train['family_memb_count']))
```

### Add potential friendship

```python
#is potential family when they have the following in common
### different surname
### same embarkation
### same ticket
### same fare

titanic_raw_train['friend_number'] = titanic_raw_train.groupby(['Ticket', 'Fare', 'Pclass', 'Embarked']).ngroup()
titanic_raw_train['friends_count'] = titanic_raw_train.groupby(['friend_number'])['family_number'].transform('count')
```

```python
has_friend_df = titanic_raw_train[(titanic_raw_train['friends_count']>1) & (titanic_raw_train['has_family']==False)]
has_friend_df['has_friend'] = True 
```

```python
titanic_raw_train = pd.concat([titanic_raw_train, has_friend_df['has_friend']], axis=1)
titanic_raw_train['has_friend'] = titanic_raw_train['has_friend'].replace(np.nan, False)
```

```python
titanic_raw_train
```

### Add single child / minor column

```python
#Single child taken to be below age 7 - age grp 0 and 1
#Single minor taken to be age 7 - 14 - age grp 2
#And no sibSp and no Parch
#And travelling without family or friends
```

```python
single_minor_df = titanic_raw_train[(titanic_raw_train['age_grp']==2) &
                                    (titanic_raw_train['SibSp']==0) &
                                    (titanic_raw_train['Parch']==0) &
                                    (~titanic_raw_train['has_friend']) &
                                    (~titanic_raw_train['has_family'])]
```

```python
single_minor_df
```

```python
single_child_df = titanic_raw_train[((titanic_raw_train['age_grp']==1) | (titanic_raw_train['age_grp']==0)) &
                                    (titanic_raw_train['SibSp']==0) &
                                    (titanic_raw_train['Parch']==0) &
                                    (~titanic_raw_train['has_friend']) &
                                    (~titanic_raw_train['has_family'])]
```

```python
single_child_df
```

```python
single_child_df_w_friend = titanic_raw_train[((titanic_raw_train['age_grp']==1) | (titanic_raw_train['age_grp']==0)) &
                                    (titanic_raw_train['SibSp']==0) &
                                    (titanic_raw_train['Parch']==0) &
                                    (~titanic_raw_train['has_family'])]
```

```python
single_child_df_w_friend
```

### drop columns that are not useful

```python
titanic_raw_train.drop(['family_number', 'friend_number'], axis=1, inplace=True)
```

### fare grouping

```python
# remove outlier
fares = titanic_raw_train[['Pclass', 'Fare']]
```

```python
fares["is_outlier"] = fares.groupby("Pclass").transform(lambda x: (x - x.mean()).abs() > 3*x.std())
```

```python
fares_wo_outliers = fares[~fares['is_outlier']]
```

```python
fares_wo_outliers["group_min_fare"] = fares_wo_outliers.groupby("Pclass")['Fare'].transform(lambda x: x.min())
```

```python
fares_wo_outliers["group_max_fare"] = fares_wo_outliers.groupby("Pclass")['Fare'].transform(lambda x: x.max())
```

```python
fares_wo_outliers
```

## cluster to find more suitable range

```python
from sklearn.cluster import DBSCAN
```

```python
clusters = DBSCAN(eps=3, min_samples=10).fit(fares_wo_outliers[['Fare']])
```
